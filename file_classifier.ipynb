```python
def classify_files(folderpath):
    import glob
    import os
    import shutil
    from glob import glob

#フォルダ内の 済.pdfのみ選択
    folder = glob( folderpath + "/*済.pdf", recursive=True)

    for file in folder:

        filename = str(file)[-23:]#ファイル名（数字済.pdf)
        year = r"/content/drive/MyDrive/東進 採点/Archives/" + filename[6:8] #年度のフォルダパス
        number = year + '/' + filename[8:10] #年度＋問題番号のフォルダパス
        #フォルダ作成(なければ)
        if not os.path.exists(year):#年度の有無確認
            os.mkdir(year)
            print('year ' + filename[6:8] + ' created')
        if not os.path.exists(number):#問題番号の有無確認
            os.mkdir(number)
            print('year ' + filename[6:8] + ' number ' + filename[8:10] + ' created')

        #移動(なければ)
        if not os.path.exists(number + '/' + filename):
            shutil.move(file, number)
            print(filename + '  moved to  ' + number)
            print('\n')
```

```python
classify_files('/content/drive/MyDrive/東進 採点/20230213')
```
